```python
import json
import pandas as pd
```

### Acquisizione N hashtag più utilizzati

```python
max_rows_hashtags=100
lower_bound_count=100
path_data_hastags=r"TGT\wordCloud.json"
path_sample_data=r"Sample\sampleWordCloud.json"
```

```python
with open(path_data_hastags, 'r') as file:
    # Load the JSON data
    data = json.load(file)


```

```python
# Sort the objects based on the "total" field
sorted_data = sorted(data.items(), key=lambda x: x[1]['total'],reverse=True)



sorted_json = {id: info for id, info in sorted_data}
```

```python
sample_hastag=dict()
i=0
for k in sorted_json.keys():
    if i<max_rows_hashtags:
        i+=1
        sample_hastag[k]=sorted_json[k]
    else:
        pass


sample_hastag
    
```

```python
with open(path_sample_data, 'w') as file:
    json.dump(sorted_json, file, indent=4)
```

### Acquisizione utenti associati agli hastag più utilizzati

```python
import ast
path_users=r"TGT\user2hastag.json"



```

```python
#acquisizione di tutti gli utenti che hanno usato quell'hastag
all_users=set()
for k in sorted_json.keys():
    values=ast.literal_eval(sorted_json[k]["users"])
    users=set(values.keys())
    all_users.update(users)
```

```python
all_users
```

```python
with open(path_users,"r") as file:
    data=json.load(file)

```

```python
#acquisisco solo i dati degli utenti presenti nella lista all_users
sample_users=dict()
for k in data.keys():
    if int(k) in all_users:
        sample_users[k]=data[k]

```

```python
with open(r"Sample\user2hastag.json","w") as f:
    json.dump(sample_users,f,indent=4)
```

### Sample Acquisizione hastag2date

```python
path_h2date=r"TGT\hashtag2date2setimentAGG.csv"
data=pd.read_csv(path_h2date)
hastags_to_get=set(sample_hastag.keys())


```

```python
output_rows=[]
cols=data.columns
for index, rows in data.iterrows():
    id=rows["Hastag"]
    if id in hastags_to_get:
        output_rows.append(list(rows))

output=pd.DataFrame(data=output_rows,columns=cols)
    
```
