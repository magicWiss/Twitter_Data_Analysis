# Word cloud

```python
import pandas as pd
import json
import pandasql
```

```python
pathData=r"SRC\tweetsAndHastags.csv"
data=pd.read_csv(pathData)
data
```

```python
class Hastag:
    def __init__(self,id,user):
        self.id=id
        self.total=1
        self.users={user:1}
        
    def to_json(self):
        return {
            "id": self.id,
            "total": str(self.total),
            "users": str(self.users)
        }   
    def update(self,user):
        self.total+=1
        if user not in self.users:
                self.users[user]=0
        self.users[user]+=1
        return self 
```

```python
word_cloud={}

autAndHastags=data[["author_id","Hastags"]]

autAndHastags
```

```python
pathWordCloud=r"TGT\wordCloud.json"

# Convert the dictionary to JSON-compatible data
json_data = {
    key: value.to_json() for key, value in word_cloud.items()
}

# Write the JSON data to a file
with open(pathWordCloud, 'w') as file:
    json.dump(json_data, file,indent=4)
```

```python
import ast
for index, row in autAndHastags.iterrows():
    
    user=row["author_id"]
    hastags=ast.literal_eval(row["Hastags"])
    for h in hastags:
        hastag=Hastag(h,user)
        if h not in word_cloud:
            word_cloud[h]=hastag
        
        else:
            word_cloud[h]=word_cloud[h].update(user)

    
   
   
word_cloud
```

```python
#ordinamento del wordCloud
tot_to_hashtag=dict()
for k in word_cloud.keys():
    total=int(word_cloud[k].total)
    
    if total not in tot_to_hashtag:
        tot_to_hashtag[total]=[]
    tot_to_hashtag[total].append(k)

myKeys = list(tot_to_hashtag.keys())
myKeys.sort(reverse=True)
sorted_dict = {i: tot_to_hashtag[i] for i in myKeys}
sorted_dict

```

```python
#create word cloud
ordered_word_cloud=dict()
for k in sorted_dict.keys():
    elements=sorted_dict[k]
    for h in elements:
        value=word_cloud[h]
        ordered_word_cloud[h]=value


```

```python

#ordinamento user map per hastag
for k in ordered_word_cloud.keys():
    obj_elm=ordered_word_cloud[k]
    ordered_set= dict(sorted(obj_elm.users.items(), key=lambda x: x[1], reverse=True))
    ordered_word_cloud[k].users=ordered_set
```

```python
pathWordCloud=r"TGT\wordCloud.json"


# Convert the dictionary to JSON-compatible data
json_data = {
    key: value.to_json() for key, value in ordered_word_cloud.items()
}

# Write the JSON data to a file
with open(pathWordCloud, 'w') as file:
    json.dump(json_data, file,indent=4)
```
