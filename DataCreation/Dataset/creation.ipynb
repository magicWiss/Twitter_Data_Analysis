# Estrazione Hastag

```python
import pandas as pd
import json
```

```python
path_src=r"SRC\qatar.csv"
data=pd.read_csv(path_src)
```

```python
tweets=data['text']
hastag_for_tweet=[]
```

```python
#Funzione estrazione hastags
import re

def extract_hashtags(tweet):
    hashtags = re.findall(r'\B#\w*[a-zA-Z]+\w*', tweet)
    out=[]
    for h in hashtags:
        
        text=h.split("#")[1]
        text=text.lower()
        
        out.append(text)
    return out
        
    
```

```python
for t in tweets:
    current_hastags=extract_hashtags(t)
    hastag_for_tweet.append(current_hastags)
hastag_for_tweet
```

```python
data['Hastags']=hastag_for_tweet
data.to_csv(r"SRC\tweetsAndHastags.csv",index=False)
```
